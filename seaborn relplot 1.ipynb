```python
taxis = sns.load_dataset('taxis')
```

```python
taxis
```

```python
taxis['payment'].unique()
```

```python
taxis['tolls'].unique()
```

```python
sns.relplot(x='total',y='tip',data=taxis)
plt.show()
```

```python
sns.relplot(x='total',y='tip',data=taxis,kind='scatter')
plt.show()
```

```python
sns.relplot(x='total',y='tip',data=taxis,size='total')
plt.show()
```

```python
sns.relplot(x='total',y='tip',data=taxis,size='total',kind='scatter',hue='color',palette='crest')
plt.show()
```

```python
sns.relplot(x='total',y='tip',data=taxis,hue='distance')
plt.show()
```

```python
sns.relplot(x='total',y='tip',data=taxis,hue='color',style='color')
```
