Refer This website for seaborn : https://towardsdatascience.com/mastering-catplot-in-seaborn-categorical-data-visualization-guide-abab7b2067af

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
exercise = sns.load_dataset("exercise")
```

```python
exercise
```

```python
exercise.info()
```

```python
exercise.columns
```

```python
 sns.catplot(x="time", y="pulse",hue="kind", data=exercise)
```

```python
sns.catplot(x="time", y="pulse",hue="kind", data=exercise,kind='point')
```

```python
sns.catplot(x="time", y="pulse",hue="kind", data=exercise,kind='bar')
```

```python
sns.catplot(x="time", y="pulse",hue="kind", data=exercise,kind='violin')
```

```python
 sns.catplot(x="time", y="pulse",hue="kind",col="diet",data=exercise)
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="alive", col="deck", col_wrap=4,data=titanic[titanic.deck.notnull()],kind="count", height=2.5, aspect=.8)
```

```python
sns.catplot(x="age", y="embark_town",hue="sex", row="class",data=titanic[titanic.embark_town.notnull()],
                orient="h", height=2, aspect=3, palette="Set3",kind="violin", dodge=True, cut=0, bw=.2)
```

```python
tips = sns.load_dataset('tips')
sns.catplot(x='day',y='total_bill',data=tips,kind='box');
```

```python
tips = sns.load_dataset('tips')
sns.catplot(x='day',y='total_bill',data=tips,kind='boxen');
```

```python
tips = sns.load_dataset('tips')
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm');
```

```python
tips = sns.load_dataset('tips')
sns.catplot(x='day',y='total_bill',data=tips,kind='strip');
```
